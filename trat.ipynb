```python
import pandas as pd
import matplotlib.pyplot as plt
import plotnine as p9
```

```python
tmin_ay=pd.read_csv('hist_tmin_mensual_ayac.csv',index_col=None)
tmax_ay=pd.read_csv('hist_tmax_mensual_ayac.csv',index_col=None)
t_ay=pd.read_csv('hist_temp_mensual_ayac.csv',index_col=None)
```

```python
tmid_ay=tmin_ay.describe().T
tmid_ay=pd.DataFrame(tmid_ay)
tmid_ay = tmid_ay.rename_axis('time').reset_index()
tmid_ay.drop([0,1,2],axis=0).drop(['count'],axis=1).plot(x='time', y='mean')
```

```python
tmad_ay=tmax_ay.describe().T
tmad_ay=pd.DataFrame(tmad_ay)
tmxd_ay = tmad_ay.rename_axis('time').reset_index()
tmxd_ay.drop([0,1,2],axis=0).drop(['count'],axis=1).plot(x='time', y='mean')
```

```python
td_ay=t_ay.describe().T
td_ay=pd.DataFrame(td_ay)
td_ay = td_ay.rename_axis('time').reset_index()
td_ay.drop([0,1,2],axis=0).drop(['count'],axis=1).plot(x='time', y='mean')
```
