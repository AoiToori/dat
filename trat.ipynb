```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Datos de ayacucho

```python
tmin_ay=pd.read_csv('https://raw.githubusercontent.com/AoiToori/dat/master/hist_tmin_mensual_ayac.csv',index_col=None)
tmax_ay=pd.read_csv('https://raw.githubusercontent.com/AoiToori/dat/master/hist_tmax_mensual_ayac.csv',index_col=None)
t_ay=pd.read_csv('https://raw.githubusercontent.com/AoiToori/dat/master/hist_temp_mensual_ayac.csv',index_col=None)
tmin_ay
```

```python
tmid_ay=tmin_ay.describe().T
tmid_ay = tmid_ay.rename_axis('time').reset_index()
tmid_ay=tmid_ay.drop([0,1,2],axis=0).drop(['count'],axis=1)
tmid_ay
```

```python
tmad_ay=tmax_ay.describe().T
tmad_ay = tmad_ay.rename_axis('time').reset_index()
tmad_ay=tmad_ay.drop([0,1,2],axis=0).drop(['count'],axis=1)
tmad_ay
```

```python
td_ay=t_ay.describe().T
td_ay = td_ay.rename_axis('time').reset_index()
td_ay =td_ay.drop([0,1,2],axis=0).drop(['count'],axis=1)
td_ay
```

```python
x=tmid_ay['time']
y3=td_ay['mean']
y1=tmad_ay['mean']
y2=tmid_ay['mean']
fig, ax=plt.subplots()
ax.plot(x,y1,x,y3,x,y2)
plt.xticks(np.arange(0, 432, step=80),fontsize=8) 
plt.legend(["T.Max","T","T.Min"],fontsize=8)
```

# Prueba de fotoperiodo


```python
x = np.arange(0,366,30)
decli = -23.45*np.cos((360*np.pi*x)/(365*180))*(np.pi/180)
hay=np.arccos(-np.tan(-13.897887*(np.pi/180))*np.tan(decli))*(180/np.pi)
hap=np.arccos(-np.tan(-14.0*(np.pi/180))*np.tan(decli))*(180/np.pi)
hpu=np.arccos(-np.tan(-13.897887*(np.pi/180))*np.tan(decli))*(180/np.pi)
nay=(2*hay)/15
nap=(2*hap)/15
fig,ax=plt.subplots()
ax.plot(x,nay,color = 'blue')
ax.plot(x,nap,color = 'red')
plt.xlabel('Tiempo(dias)')
plt.ylabel('horas')
plt.title('Fotoperiodo')
plt.legend(["Latitud 0","Latitud 12","Latitud 30","Latitud 45","Latitud 60"])
plt.grid(True)
plt.show()
```
